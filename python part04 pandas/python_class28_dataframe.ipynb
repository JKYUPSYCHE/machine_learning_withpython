#### NaN 처리 방법
- 데이터에서 삭제 : dropna - NaN이 존재하는 row를 삭제
- 다른 값으로 치환 : fillna

- NaN값 삭제하기

```python
train_data.dropna()  # NaN이 존재하는 행(row)을 삭제
```

```python
train_data.dropna(subset = ['Age'])  # 특정 column의 NaN이 있다면 그 row를 삭제
```

```python
train_data.dropna(axis = 1)  # NaN이 있는 column을 삭제
```

- NaN값 치환하기
  - 평균으로 대체하기
  - 생존자/사망자 별 평균으로 대체하기

```python
train_data['Age'].fillna(train_data['Age'].mean())  # Age의 NaN을 Age의 평균값으로 치환
```

```python
# 생존자 나이 평균
mean0 = train_data[train_data['Survived'] == 1]['Age'].mean()

# 사망자 나이 평균
mean1 = train_data[train_data['Survived'] == 0]['Age'].mean()
print(mean0, mean1)
```

```python
train_data[train_data['Survived'] == 1]['Age'].fillna(mean1)
train_data[train_data['Survived'] == 0]['Age'].fillna(mean0)
```

```python
train_data.loc[train_data['Survived'] == 1, 'Age'] = train_data[train_data['Survived'] == 1]['Age'].fillna(mean1)
train_data.loc[train_data['Survived'] == 0, 'Age'] = train_data[train_data['Survived'] == 0]['Age'].fillna(mean0)
```

#### 범주형(categorical type) 데이터
- 연속적이지 않은 값을 갖는 데이터
  - Name, Sex, Ticket, Cabin, Embarked
- 어떤 경우, 숫자형 타입이라 할지라도 개념적으로 범주형으로 처리해야할 경우가 있음
  - ex> Pclass - 클래스를 구분하는 역할을 하므로 범주형 자료형으로 간주

```python
train_data.info()
```

```python
train_data['Pclass'] = train_data['Pclass'].astype(str)
```

```python
train_data.info()
```

- 변환 로직을 함수로 만든 후, apply 함수로 적용

```python
import math
```

```python
def age_categorize(age):
    if math.isnan(age):  #  NaN 처리
        return -1
    return math.floor(age/10) * 10  # math 모듈의 floor를 통해 소수점을 버림
```

```python
train_data['Age'].apply(age_categorize)
```

#### 범주형 데이터 전처리 하기

#### one-hot encoding
- 범주형 데이터는 분석단계에서 계산이 어렵기 때문에 숫자형으로 변경이 필요
- 범주형 데이터의 각 범주를 column레벨로 변경
- 해당 범주에 해당하면 1, 아니면 0으로 채우는 인코딩 기법
- pandas.get_dummies 함수 사용
  - drop_first : 첫번쩨 카테고리 값은 사용하지 않음

```python
pd.get_dummies(train_data, columns = ['Pclass', 'Sex', 'Embarked'], drop_first = False)
```

```python
pd.get_dummies(train_data, columns = ['Pclass', 'Sex', 'Embarked'], drop_first = True)
```
