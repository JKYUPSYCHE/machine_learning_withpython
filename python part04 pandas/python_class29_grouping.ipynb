```python
import pandas as pd
import numpy as np
```

### group by
- 아래의 세 단계를 적용하여 데이터를 그룹화(grouping)
  - 데이터 분할
  - operation 적용
  - 데이터 병합

```python
df = pd.read_csv('./train.csv')
df.head()
```

#### groupby groups 속성
- 각그룹과 그룹에 속한 index를 dict 형태로 표현

```python
class_group = df.groupby('Pclass')
class_group
```

```python
class_group.groups
```

```python
gender_group = df.groupby('Sex')
gender_group.groups
```

#### grouping 함수
- 그룹 데이터에 적용 가능한 통계 함수**(NaN은 제외)**
- count : 데이터 개수
- sum : 데이터의 합
- mean, std, var : 평균, 표준편차, 분산
- min, max : 최소, 최대값

```python
class_group.count()
```

```python
class_group.mean()['Survived'] # 클래스에 따른 생존률
```

```python
gender_group.mean()  # 남녀별 생존률을 확인하고자 할 때
```

```python
df.groupby(['Pclass', 'Sex']).mean()
```

```python
df.groupby(['Pclass', 'Sex']).mean().loc[(2, 'female')]  # plcass가 2, female인 row의 평균값만 가져옴
```

#### index를 이용한 group by
- index가 있는 경우, groupby 함수에 level 사용 가능
  - level은 index의 depth를 의미하며, 가장 왼쪽부터, 0부터 증가

- set_index 함수
  - coiumn 데이터를 index 레벨로 변경
- reset_index 함수
  - 인덱스 초기화

```python
df.head()
```

```python
df.set_index(['Pclass', 'Sex'])  # 멀티 인덱스로도 가능
```

```python
df.set_index(['Pclass', 'Sex']).reset_index()  # 0 based index로 초기화
```

```python
df.set_index('Age').groupby(level = 0).mean()
```

#### 나이대별 생존율 구하기

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10) * 10
```

```python
df.set_index('Age').groupby(age_categorize).mean()['Survived']
```

####  Multiindex를 이용한 grouping

```python
df.set_index(['Pclass', 'Sex']).groupby(level = 0).mean() # Pclass가 level 0 Sex가 1
```

#### aggregate 함수 사용하기
- groupby 결과에 집계함수를 적용하여 그룹별 데이터 확인 가능

```python
df.set_index(['Pclass', 'Sex']).groupby(level = [0, 1]).aggregate([np.mean, np.sum, np.max])
```

### transform 함수
- groupby후 transform 함수를 사용하면 원래의 index를 유지한 상태로 통계함수를 적용
- 전체 데이터의 집계가 아닌 각 그룹에서의 집계를 계산
- 따라서 새로 생성된 데이터를 원본 datafrane과 합치기 쉬움

```python
df.groupby('Pclass').mean()
```

```python
df.groupby('Pclass').transform(np.mean)
```

```python
df['Age2'] = df.groupby('Pclass').transform(np.mean)['Age']
df
```

```python
df.groupby(['Pclass', 'Sex']).mean()
```

```python
df.groupby(['Pclass', 'Sex']).transform(np.mean)
```

### pivot 테이블의 이해

```python
df = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
                  })

df
```

#### pivot
- dataframe의 형태를 변경
- index, columnm, data로 사용할 column을 명시

```python
df.pivot('지역', '요일')  # 중복을 허용하지 않는다. Ex 서울의 월요일이 2번이면 ERROR
```

#### pivot_table
- 기능적으로 pivot과 동일
- pivot과의 차이점
  - 중복되는 모호한 값이 있을 경우, aggregation 함수를 사용하여 값을 채움

```python
pd.pivot_table(df, index = '요일', columns = '지역', aggfunc = np.mean)
```

#### stack & unstack
- stack : 컬럼 레벨에서 인덱스 레벨로 dataframe 변경
  - 즉, 데이터를 쌓아올리는 개념으로 이해하면 쉬움
- unstack : 인덱스 레벨에서 컬럼 레벨로 dataframe 변경
  - stack의 반대 operation
- 둘은 역의 관계에 있음

```python
new_df = df.set_index(['지역', '요일'])
new_df
```

```python
new_df.unstack(0)  # 지역 인덱스를 column 레벨로 올려라
```

```python
new_df.unstack(1)
```

```python
new_df.unstack(0).stack(0)
```

```python
new_df.unstack(0).stack(1)
```

#### concat 함수를 사용하여 dataframe 병합하기
- pandas.concat 함수
- 축을 따라 dataframe을 병합 가능
  - 기본 axis = 0 -> 행단위 병합

- column 명이 같은 경우

```python
df1 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
```

```python
df2 = pd.DataFrame({'key1' : np.arange(10), 'value1' : np.random.randn(10)})
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1, df2], ignore_index = True)  # 새로운 index 생성
```

```python
pd.concat([df1, df2], axis = 1)  # column level 병함. default = 0
```

- column 명이 다른 경우

```python
df3 = pd.DataFrame({'key2' : np.arange(10), 'value2' : np.random.randn(10)})
```

```python
pd.concat([df1, df3])
```

### dataframe merge
- SQL의 join처럼 특정한 column을 기준으로 병합
  - join 방식 : how 파라미터를 통해 명시
    - inner : 기본값, 일치하는 값이 있는 경우
    - left : left outer join
    - right : right outer join
    - outer : full outer join
  - pandas.merge 함수가 사용됨

```python
customer = pd.DataFrame({'customer_id' : np.arange(6), 
                    'name' : ['철수'"", '영희', '길동', '영수', '수민', '동건'], 
                    '나이' : [40, 20, 21, 30, 31, 18]})

customer
```

```python
orders = pd.DataFrame({'customer_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9], 
                    'item' : ['치약', '칫솔', '이어폰', '헤드셋', '수건', '생수', '수건', '치약', '생수', '케이스'], 
                    'quantity' : [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
orders.head()
```

- on
  - join 대상이 되는 column 명시

```python
pd.merge(customer, orders, on='customer_id', how = 'inner')  #  how 파라미터를 통해 방식 설정. default는 inner
```

```python
pd.merge(customer, orders, on='customer_id', how = 'right')
```

```python
pd.merge(customer, orders, on='customer_id', how = 'outer')
```

- index 기준으로 join

```python
cust1 = customer.set_index('customer_id')
order1 = orders.set_index('customer_id')
```

```python
cust1
```

```python
order1
```

```python
pd.merge(cust1, order1, left_index = True, right_index = True)
```
